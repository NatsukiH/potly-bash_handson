```python
# マイクロソフト社の株価データ

import yfinance as yf

msft = yf.Ticker("MSFT")
hist = msft.history(period="1y")
hist.head()
```

### Q1

indexをX, Close列をYにとった折れ線グラフを描画してください.

ヒント:  
indexは `hist.index` , Close列は `hist["Close"])` で取得できます.

![](https://github.com/plotly-dash-book/events/blob/main/20201205/assets/plotly_py_q1.png?raw=true)

```python
import plotly.graph_objects as go

fig = go.Figure()
scatter_trace = go.Scatter(x=hist.index, y=hist["Close"])
go.Figure(data=scatter_trace).show()
```

### Q2

前の折れ線グラフと, indexをX, Volume列をYにとった棒グラフを縦に並べてください.

![](https://github.com/plotly-dash-book/events/blob/main/20201205/assets/plotly_py_q2.png?raw=true)

```python
bar_trace = go.Bar(x=hist.index, y=hist["Volume"])
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=2)
fig.add_trace(scatter_trace, row=1, col=1)
fig.add_trace(bar_trace, row=2, col=1)
```

> 「p81 4.2.3 軸の設定」より

範囲の異なるデータを同一のサブプロットに描画し， 2 軸目を右（Y 軸）または上（X 軸）に配置するには次の手順を実施します．次の手順は Y 軸の設定方法です． X 軸を設定する場合は `secondary_y` を `secondary_x` に置き換えます．

1. `make_subplots` 関数の引数 specs の要素に `{"secondary_y": True}` を渡す（❶）
2. figure の `add_trace` メソッドの引数 `secondary_y` に `True` を渡す（❷）
3. figure の `update_yaxes` メソッドの引数 `secondary_y` に True を渡す（❸）

次のコードでは異なる範囲の Y 値の折れ線グラフを同一のサブプロットに描画しています．

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Y軸に2軸目が存在するサブプロット
two_yaxis_fig = make_subplots(specs=[[{"secondary_y": True}]]) 
two_yaxis_fig.add_trace(go.Scatter(x=[1, 2, 3], y=[3, 2, 4], name="1st"))
two_yaxis_fig.add_trace(
    go.Scatter(x=[1, 2, 3], y=[1, 20, 15], name="2nd"),
    secondary_y=True,  # traceを2軸目とする
)
two_yaxis_fig.update_yaxes(title={"text": "1st"}, showgrid=False)
two_yaxis_fig.update_yaxes(
    secondary_y=True,  # 2軸目の設定
    title={"text": "2nd"},  # 軸ラベルの指定
    showgrid=False  # 補助線を非表示
)
two_yaxis_fig.show()
```

### Q2

前の折れ線グラフと棒グラフを1つのグラフに描画してください.

![](https://github.com/plotly-dash-book/events/blob/main/20201205/assets/plotly_py_q3.png?raw=true)

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Y軸に2軸目が存在するサブプロット
two_yaxis_fig = make_subplots(specs=[[{"secondary_y": True}]]) 
two_yaxis_fig.add_trace(bar_trace)
two_yaxis_fig.add_trace(
    scatter_trace,
    secondary_y=True,  # traceを2軸目とする
)
two_yaxis_fig.update_yaxes(
    secondary_y=True,  # 2軸目の設定
)
two_yaxis_fig.show()
```
