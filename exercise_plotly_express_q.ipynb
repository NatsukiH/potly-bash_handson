## Plotly Express 演習問題

```python
import plotly.express as px

tips = px.data.tips()
tips.head()
```

列名|-
---|---
total_bill|会計金額
tip|チップの支払額
sex|性別
smoker|喫煙の有無
day|曜日
time|時間帯(ランチ/ディナー)
size|テーブルあたりの客数

### Q1

total_bill列をX, tip列をYにとった散布図を描画してください.

![](https://github.com/plotly-dash-book/events/blob/main/20201205/assets/plotly_express_q1.png?raw=true)

```python
px.scatter(tips, x="total_bill", y="tip").show()
```

### Q2

前の散布図から要素の大きさをsize列の値にしてください.

![](https://github.com/plotly-dash-book/events/blob/main/20201205/assets/plotly_express_q2.png?raw=true)

```python
px.scatter(
    tips, 
    x="total_bill", 
    y="tip", 
    size="size",  # sizeを点の大きさで表現
    size_max=30,  # 点の大きさの最大値)
).show()
```

### Q3

前の散布図からday列で色分けしてください.

![](https://github.com/plotly-dash-book/events/blob/main/20201205/assets/plotly_express_q3.png?raw=true)

```python
px.scatter(
    tips, 
    x="total_bill", 
    y="tip", 
    size="size",  # sizeを点の大きさで表現
    size_max=30,  # 点の大きさの最大値)
    color="day"
).show()
```

### Q4

前の散布図からtime列で縦(行)に分割し, smoker列で横(列)に分割してください

![](https://github.com/plotly-dash-book/events/blob/main/20201205/assets/plotly_express_q4.png?raw=true)

```python
px.scatter(
    tips, 
    x="total_bill", 
    y="tip", 
    size="size",  # sizeを点の大きさで表現
    size_max=30,  # 点の大きさの最大値)
    color="day",
    facet_row="time",
    facet_col="smoker",
).show()
```
